```python
import pandas as pd
from bokeh.plotting import show, figure, output_notebook
from bokeh.models import ColumnDataSource
# get data first
df = pd.read_csv('../rawData/^GDAXI.csv', parse_dates=['Date'], infer_datetime_format=True, index_col='Date')
df = df.fillna(0)
dax_cds = ColumnDataSource(df)
# output_file('my_first_graph.html')
p = figure(x_axis_type="datetime",plot_width=1500, plot_height=400,)
p.vbar(x='Date', top='Close', bottom='Open', width=1000, color='Blue', legend='DAX', source=dax_cds)
output_notebook()
# show(p)

```

```python
show(p)
```

```python
df.Close.apply([:+14].max()
```

```python
everincreasing = df.Close[df.Close.cummax().pct_change() > 0 ]
everincreasing
```

```python
selldates = df.Close[df.Close.cummax().pct_change() > 0 ].index.to_series().diff().ge('14 days')
everincreasing[selldates]
p.line(everincreasing[selldates].index, everincreasing[selldates], color = 'Green')
output_notebook()
show(p)
```
